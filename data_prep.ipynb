```python
from datasets import load_dataset
```

```python
dataset = load_dataset("ai4bharat/indic-instruct-data-v0.1", "anudesh", split= "hi")
print(dataset)
```

```python
{"messages" : dataset[0]['messages']}
```

```python
jsonl_data = []
```

```python
for data in dataset:
    req_format = {"messages" : data['messages']}

    jsonl_data.append(req_format)
```

```python
jsonl_data[:3]
```

```python
import jsonlines

output_file_path = "data/hindi_data.jsonl"

with jsonlines.open(output_file_path, mode="w") as writer:
    writer.write_all(jsonl_data)
```
